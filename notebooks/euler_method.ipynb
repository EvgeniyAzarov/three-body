## Setting

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
plt.rcParams.update({"figure.figsize": (20, 6)})
```

## Euler method

$$x' = f(x, t), \quad x(t_0) = x_0$$

$$x_{n+1} = x_n + \Delta t \cdot x'(t_n) = x_n + \Delta t \cdot f(x_n, t_n)$$

```python
def euler_solver(f, dt, t0, x0, n):
    t = np.arange(t0, t0 + n * dt, dt)

    x = [x0]
    for i in range(1, len(t)):
        x.append(x[i - 1] + dt * f(t[i - 1], x[i - 1]))

    return t, np.array(x)
```

## Simple test

```python
def f(t, x):
    return 2 * t
```

```python
t1, x1 = euler_solver(f, 0.1, 0, 0, 10)
t2, x2 = euler_solver(f, 0.01, 0, 0, 100)
```

```python
sol = sci.integrate.solve_ivp(f, (0, 1), (0,), t_eval=np.linspace(0, 1, 100))
```

```python
plt.plot(t2, t2 ** 2, label='analytical solution')

plt.plot(t1, x1, label='dt=0.1')
plt.plot(t2, x2, label='dt=0.01')
plt.plot(sol.t, sol.y[0])

plt.legend()
plt.show()
```

## More complicated

```python
plt.plot(t, np.exp(t))
plt.plot(t, x)
plt.show()
```
